# Analyze Sales Data of Electronic Store

## Problem Statement:

This project aims to analyze sales data of Electronic Store to derive valuable insights for business decision-making. 
The objectives are as follows:

* **Analyzing Monthly Sales**: Explore the monthly sales trends to identify patterns and fluctuations in sales volume over time.

* **Identifying the City with Maximum Orders**: Determine which city has placed the highest number of orders to target marketing efforts and optimize logistics.

* **Determining Top-Selling Products and Reasons**: Identify the best-selling product and investigate the factors contributing to its success.

* **Understanding Trend of Top-Selling Product**: Analyze the sales trend of the most sold product to understand its popularity dynamics and anticipate future demand.

* **Analyzing Product Bundling**: Identify which products are frequently sold together to optimize product bundling strategies and enhance cross-selling opportunities.

By addressing these objectives, this project aims to provide actionable insights that can drive strategic decisions to boost sales, optimize inventory management, and enhance overall business performance.







## Electronic Store Dataset

Following Columns:

* **Order ID** : This column contains unique identifiers or codes for each order. These IDs are typically used to uniquely identify and track individual orders.

* **Product** : This column indicates the name or description of the product that was ordered as part of the transaction.

* **Quantity Ordered** : This column represents the quantity of the product ordered in each transaction.

* **Price Each** : This column contains the price of each unit of the product.

* **Order Date** : This column indicates the date and time when the order was placed.

* **Purchase Address** : This column provides the address where the order was shipped or delivered.

### Analyzing Monthly Sales

```python
Sales_data.info()
```

```python
Sales_data['Quantity Ordered'] = pd.to_numeric(Sales_data['Quantity Ordered'], errors='coerce')
```

```python
# Change Datatype
Sales_data['Quantity Ordered'] = Sales_data['Quantity Ordered'].astype(int)
Sales_data['Price Each'] = Sales_data['Price Each'].astype(float)
Sales_data['Order Date'] = Sales_data['Order Date'].astype(np.datetime64)
```

```python
Sales_data.info()
```

```python
# Create new column 
Sales_data['Total Sale'] = Sales_data['Quantity Ordered']*Sales_data['Price Each']
```

```python
Sales_data
```

```python
Sales_data['Month'] = Sales_data['Order Date'].dt.month
```

```python
Sales_data
```

```python
Sales_data['Month'].unique()
```

```python
Sales_data.shape
```

```python
Sales_data.groupby(['Month'])['Total Sale'].sum()
```

```python
plt.figure(figsize=(12, 6));
Sales_data.groupby(['Month'])['Total Sale'].sum().plot(kind = 'bar')
plt.xlabel('Months', fontsize = 14)  
plt.ylabel('Total Sales in Million', fontsize = 14)
```

    Last month of the year has highest sale.

    Furthermore, the sales data displays a cyclic pattern over the course of the year.

### Identifying the City with Maximum Orders

```python
Sales_data
```

```python
Sales_data['City'] = Sales_data['Purchase Address'].str.split(',').str.get(1)
```

```python
Sales_data
```

```python
city_order_counts = Sales_data.groupby(['City'])['Quantity Ordered'].count().reset_index()
```

```python
city_order_counts
```

```python
city_order_counts_sorted = city_order_counts.sort_values(by='Quantity Ordered', ascending=False)
```

```python
city_order_counts_sorted
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(15, 7))
ax = sns.barplot(x='Quantity Ordered', y='City', data=city_order_counts_sorted, palette='viridis')


plt.xlabel('Number of Orders')
plt.ylabel('City')
plt.title('Orders by City')


for p in ax.patches:
    width = p.get_width()
    plt.text(width, p.get_y() + p.get_height() / 2, f'{int(width)}', ha="left", va="center")


plt.show()
```

    San Francisco and Los Angeles Placed the maximum orders of 45k and 30k respectively.

### Determining Top-Selling Products and Reasons

```python
Sales_data
```

```python
count_df = Sales_data.groupby(['Product']).agg({'Quantity Ordered': 'sum', 'Price Each': 'mean'})
```

```python
count_df = count_df.reset_index()
```

```python
count_df
```

```python
products = count_df['Product'].values
```

```python
fig, ax1 = plt.subplots(figsize = (15,8))  

ax2 = ax1.twinx()

ax1.bar(count_df['Product'], count_df['Quantity Ordered'], color = 'g')
ax2.plot(count_df['Product'], count_df['Price Each'])
ax1.set_xticklabels(products, rotation = 'vertical', fontsize=12)

ax1.set_ylabel('Order Count')
ax2.set_ylabel('Avg Price of product')
```

    The highest-selling products are "4-packs of AA batteries" and "4-packs of AAA batteries".
    Additionally, there exists an inverse relationship between the average price and the number of orders, 
    meaning that when the average price is lower, the frequency of orders tends to be higher.

### Understanding Trend of Top-Selling Product

```python
Sales_data
```

```python
Sales_data.groupby('Product').agg({'Quantity Ordered':'count'}).sort_values( by = 'Quantity Ordered', ascending = False).nlargest(5, 'Quantity Ordered').index
```

```python
most_sold_product = Sales_data.groupby('Product').agg({'Quantity Ordered':'count'}).sort_values( by = 'Quantity Ordered', ascending = False).nlargest(5, 'Quantity Ordered').index
```

```python
most_sold_product_df = Sales_data[Sales_data['Product'].isin(most_sold_product)]
```

```python
pivot = most_sold_product_df.pivot_table(values = 'Quantity Ordered', index = 'Month', columns = 'Product', aggfunc = 'count')
```

```python
pivot
```

```python
pivot.plot(figsize = (15,10))
```

    The leading products experience periodic fluctuations in demand throughout the year, 
    with the highest sales occurring at 4th and 12th month.

### Analyzing Product Bundling

```python
Sales_data
```

```python
df_duplicated = Sales_data[Sales_data['Order ID'].duplicated(keep = False)]
```

```python
df_duplicated
```

```python
duplicate_products = df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x : ','.join(x)).reset_index().rename(columns = {'Product': 'Grouped_Products'})
```

```python
duplicate_products
```

```python
dup_prod_df = df_duplicated.merge(duplicate_products, how='left', on='Order ID')
```

```python
dup_prod_df
```

```python
no_dup_df = dup_prod_df.drop_duplicates(subset='Order ID')
```

```python
no_dup_df
```

```python
product_counts =no_dup_df['Grouped_Products'].value_counts()[0:5].reset_index()
```

```python
product_counts
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(15, 8))
ax = sns.barplot(x='Grouped_Products', y='index', data=product_counts, palette='viridis')


plt.xlabel('Count')
plt.ylabel('Products')
plt.title('Top 5 Grouped Products')


for p in ax.patches:
    width = p.get_width()
    plt.text(width, p.get_y() + p.get_height() / 2, f'{int(width)}', ha="left", va="center")


plt.show()
```

    It is beneficial to utilize product recommendations to enhance the shopping experience. 
    For instance, if a customer has purchased an iPhone, it is highly probable that they may also be interested in acquiring a Lightning charging cable.
    Therefore, it is advisable to suggest this complementary product to the customer.
